```python
from autoviz.AutoViz_Class import AutoViz_Class

AV = AutoViz_Class()
```

```python
filename = "/content/train.csv"
sep = ","
dft = AV.AutoViz(
    filename,
    sep=",",
    depVar="",
    dfte=None,
    header=0,
    verbose=1,
    lowess=False,
    chart_format="svg",
    max_rows_analyzed=150000,
    max_cols_analyzed=30,
)
```
